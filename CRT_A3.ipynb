```python
''' code to claculate the area of a circle by importing pi from from math and create a
class for circle with a constructor, whrere radius is considered as an arg
calculate and return the value to the object .'''

import math
class circle:
    def __init__(self, radius):
        self.radius= radius
    def area(self):
        return math.pi * self.radius**2

    def cir(self):
        return 2*math.pi*self.radius


r= int(input("enter the radius"))
c= circle(r)
print(f"area of circle :{c.area():.2f}")
print("circumference of circle :", c.cir())
```

```python
''' code to find and print the factorial of a given number using a class called math
and register the given number  in a  constructor to calculate .
 return the value by passing to the main function'''

class math:
    def __init__(self,num):
        self.num=num
    def fact(self):
        f=1
        for i in range(1, self.num+1):
            f*=i
        return f
number= int(input("Enter the number"))
m= math(number)
print(f"factorial of {number} is {m.fact()}")
```

```python
''' code to calc area and circumference of a circle by using a class
rectangle , create a constructor and separate functions for area circ and delete the
constructor , import math pi value '''

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius**2

    def circumference(self):
        return 2 * math.pi * self.radius

# Example usage:
radius_value = 6
circle_instance = Circle(radius_value)
print(f"Area: {circle_instance.area()}")
print(f"Circumference: {circle_instance.circumference()}")
```

```python
# math evaluation on get item selection

''' code to add the sum of rows in a 2-d list '''

class RowSum:
    def __init__(self,matrix):
        self.matrix=matrix
    def __getitem__(self,row):
        return sum(self.matrix[row])
    def __setitem__(self,row,new_row):
        self.matrix[row]=new_row

m= RowSum([[1,2],[3,4],[5,6]])
# sum of oth index row values - 3
# sum of 2th index row values - 11

print("sumof row 0:", m[0])
print("sumof row 2:", m[2])
m[1]=[10,20]


# sum of 1th index row values, which are new = 30
print("sumof new_row :", m[1])
```

```python
''' create a class atm and check the pin user1234
define the function to withdraw 2000/- ,balance amount from
account 5999/- , if the amount is over the balance exit and check the
account for accessing the account ... '''


class atm:
    def __init__(self, pin, balance=5999):
        self.pin = pin
        self.balance = balance

    def check_pin(self, entered_pin):
        if entered_pin == self.pin:
            print("PIN accepted.")
            return True
        else:
            print("Incorrect PIN.")
            return False

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance.")
            return False
        elif amount > 2000:
            print("Withdrawal amount exceeds limit of 2000/-")
            return False
        else:
            self.balance -= amount
            print(f"Withdrawal successful. Remaining balance: {self.balance}/-")
            return True

    def check_balance(self):
        print(f"Current balance: {self.balance}/-")

user_pin = "user1234"
my_atm = atm(user_pin)

entered_pin = input("Enter your PIN: ")
if my_atm.check_pin(entered_pin):
    my_atm.check_balance()
    withdraw_amount = int(input("Enter amount to withdraw (max 2000/-): "))
    my_atm.withdraw(withdraw_amount)
```

```python
class op:
      @staticmethod
      def add(a,b):
          return a+b
      @staticmethod
      def sub(a,b):
          return a-b
      @staticmethod
      def mul(a,b):
          return a*b
      @staticmethod
      def div(a,b):
          return a/b
x=int(input("enter the value of x :"))
y=int(input("enter the value of y :"))
print("sum----", op.add(x,y))
print("sub----", op.sub(x,y))
print("mul----", op.mul(x,y))
print("div----", op.div(x,y))
```
